# Espacios de color

En este notebook veremos brevemente c√≥mo manipular espacios de color tanto en `OpenCV` como en `scikit-image`.

## OpenCV

Empecemos por cargar una imagen:

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./crayons.jpeg', cv2.IMREAD_COLOR)
print(img.shape)

# Tenemos que convertir a espacio RGB, porque OpenCV carga las im√°genes como BGR.
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
# Mostremos cada uno de los canales de la imagen en RGB.
red_channel = rgb_image[:, :, 0]
green_channel = rgb_image[:, :, 1]
blue_channel = rgb_image[:, : , 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
ax1.set_title("Rojo")
ax1.imshow(red_channel, cmap="gray")

ax2.set_title("Verde")
ax2.imshow(green_channel, cmap="gray")

ax3.set_title("Azul")
ax3.imshow(blue_channel, cmap="gray")
```

```python
# Pasemos a espacio HSV

hsv_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)

print(hsv_image.shape)

plt.imshow(hsv_image)
```

Dado que la imagen se encuentra en espacio HSV, los colores se ven distorsionados. ¬øPor qu√©? Pues, la funci√≥n imshow asume que las im√°genes vienen en formato RGB, por lo que al ser interpretadas en este espacio, obtenemos el resultado mostrado anteriormente.

Veamos c√≥mo se ve cada canal.

```python
hue_channel = hsv_image[:, :, 0]
saturation_channel = hsv_image[:, :, 1]
value_channel = hsv_image[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
ax1.set_title("Matiz")
ax1.imshow(hue_channel, cmap="gray")

ax2.set_title("Saturaci√≥n")
ax2.imshow(saturation_channel, cmap="gray")

ax3.set_title("Brillo")
ax3.imshow(value_channel, cmap="gray")
```
