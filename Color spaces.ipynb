# Color spaces

In this notebook we'll see briefly how to manipulate color spaces in both `OpenCV` and `scikit-image`.

## OpenCV

Let's start by loading an image:

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('./crayons.jpeg', cv2.IMREAD_COLOR)
print(img.shape)

# We must convert to RGB space, because OpenCV loads an image as BGR.
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
# Let's show each channel.
red_channel = rgb_image[:, :, 0]
green_channel = rgb_image[:, :, 1]
blue_channel = rgb_image[:, : , 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
ax1.set_title("Red")
ax1.imshow(red_channel, cmap="gray")

ax2.set_title("Green")
ax2.imshow(green_channel, cmap="gray")

ax3.set_title("Blue")
ax3.imshow(blue_channel, cmap="gray")
```

```python
# Let's switch to HSV space

hsv_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)

print(hsv_image.shape)

plt.imshow(hsv_image)
```

Given the image is in HSV space, the colors are distorted. Why? Well, the imshow function assumes images are in RGB, so it'll interpret each pixel as a (red, green, blue) tuple. 

Let's see how each channel looks like.

```python
hue_channel = hsv_image[:, :, 0]
saturation_channel = hsv_image[:, :, 1]
value_channel = hsv_image[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
ax1.set_title("Hue")
ax1.imshow(hue_channel, cmap="gray")

ax2.set_title("Saturation")
ax2.imshow(saturation_channel, cmap="gray")

ax3.set_title("Value")
ax3.imshow(value_channel, cmap="gray")
```

## scikit-image

Let's repeat the process using `skimage`.
